```python
def write_api_response_to_file(response, output_file_name="response.json"):
    with open(output_file_name, 'w') as f:
        json.dump(response, f, indent=4)
```

```python
# make API call to GET Channel detials
next_page_token = ""         # Find a way to loop this part of the code so 'nextPageToken' can be fetched each time

url_channel_call = "https://www.googleapis.com/youtube/v3/search?key="+API_KEY+"&channelId="+CHANNEL_ID+"&part=snippet,id&order=date&maxResults=1000000&pageToken="+next_page_token
response_channel_stats = requests.get(url_channel_call).json()   # Always helpful to save response in text file to help debug
time.sleep(1)
write_api_response_to_file(response_channel_stats, "response_channel_stats.json")

for video in response_channel_stats["items"]:
        
    video_id = video["id"]["videoId"]
    video_title = video["snippet"]["title"]
    upload_date = video["snippet"]["publishedAt"].split("T")[0]

    # make API call to GET Video detials
    url_video_details_call = "https://www.googleapis.com/youtube/v3/videos?id="+video_id+"&part=statistics&key="+API_KEY
    response_video_stats = requests.get(url_video_details_call).json()
    write_api_response_to_file(response_video_stats, "response_video_stats.json")     # After code finishes running, this file will have the json data for the last video in 'response_channel_stats' ie 50th video

    view_count = response_video_stats["items"][0]["statistics"]["viewCount"]
    like_count = response_video_stats["items"][0]["statistics"]["likeCount"]
    comment_count = response_video_stats["items"][0]["statistics"]["commentCount"]

    df = df.append({                            # Classic mistake is to not use df = df.append() and reassign the df each time. Otherwise you'll end up with empty DF...
        'video_id': video_id,
        'video_title': video_title,
        'upload_date': upload_date,
        'view_count': view_count,
        'like_count': like_count,
        'comment_count': comment_count,
    }, ignore_index=True)
```
